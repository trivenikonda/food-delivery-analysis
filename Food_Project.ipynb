```python
import pandas as pd

orders = pd.read_csv("orders.csv")
users = pd.read_json("users.json")

print(orders.head())
print(users.head())
```

```python
import sqlite3

conn = sqlite3.connect("restaurants.db")

with open("restaurants.sql", "r") as f:
    sql_script = f.read()

conn.executescript(sql_script)

restaurants = pd.read_sql_query("SELECT * FROM restaurants", conn)
print(restaurants.head())
```

```python
df = orders.merge(users, on="user_id", how="left")
df = df.merge(restaurants, on="restaurant_id", how="left")

print(df.head())
print(df.shape)
```

```python
df.to_csv("final_food_delivery_dataset.csv", index=False)
print("final_food_delivery_dataset.csv created")
```

```python
df = df.drop(columns=["restaurant_name_x"])
df = df.rename(columns={"restaurant_name_y": "restaurant_name"})
df.head()
```

```python
df["order_date"] = pd.to_datetime(df["order_date"], dayfirst=True)
df["quarter"] = df["order_date"].dt.quarter

df.head()
```

```python
q1 = df[df["membership"] == "Gold"].groupby("city")["total_amount"].sum()
q1.sort_values(ascending=False)
```

```python
df.groupby("cuisine")["total_amount"].mean().sort_values(ascending=False)
```

```python
user_spend = df.groupby("user_id")["total_amount"].sum()
(user_spend > 1000).sum()
```

```python
bins = [3.0, 3.5, 4.0, 4.5, 5.0]
labels = ["3.03.5","3.64.0","4.14.5","4.65.0"]

df["rating_range"] = pd.cut(df["rating"], bins=bins, labels=labels, include_lowest=True)

df.groupby("rating_range")["total_amount"].sum().sort_values(ascending=False)
```

```python
df[df["membership"]=="Gold"].groupby("city")["total_amount"].mean().sort_values(ascending=False)
```

```python
pd.DataFrame({
    "restaurants": df.groupby("cuisine")["restaurant_id"].nunique(),
    "revenue": df.groupby("cuisine")["total_amount"].sum()
}).sort_values("restaurants")
```
